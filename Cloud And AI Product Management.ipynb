```python
import numpy as np
import matplotlib.pyplot as plt

def create_spider_chart(categories, values, title="Product Management Primary Focus"):
    N = len(categories)

    # Create an angle array for each category
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    values += values[:1]  # Repeat first value at end for full loop
    angles += angles[:1]  # Repeat first angle

    # Create the plot
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))  # Increased figure size

    # Plot data
    ax.fill(angles, values, color='blue', alpha=0.3)
    ax.plot(angles, values, color='blue', linewidth=2)

    # Set labels
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=8, ha='center', va='top', wrap=True)  # Smaller font, centered

    # Rotate labels for better readability
    for label, angle in zip(ax.get_xticklabels(), angles[:-1]):
        label.set_rotation(np.degrees(angle) - 90)
        label.set_verticalalignment('bottom' if angle < np.pi else 'top')

    # Adjust y-axis labels
    ax.set_yticklabels([])  # Hide radial labels
    ax.set_ylim(0, 10)

    # Title
    plt.title(title, fontsize=14, fontweight='bold', pad=30)

    # Show plot
    plt.show()

# Define categories and their corresponding skill levels
categories = [
    "Financials (Cash Flow, ROI, NPV, Payback)",
        "                               Product Roadmap (Pricing, Strategic Planning, Monetization)",

    "                       AI Analytical (Product Insights, Business Intelligence, AI, Revenue Forecasting)",
    "Strategy (Alignment, Regulations, Resource Allocation)",
    "Generic (Competition, Market Research, Product KPIs)",
        "Marketing (Sales, Customer Focus)",
]

# Example skill level values (scale 0-10)
values = [8, 9, 7, 8, 9, 7]  # Adjust these based on specific assessments

# Generate the radar chart
create_spider_chart(categories, values)
```
